```python
import numpy as np
import scipy as sp
import scipy.ndimage as ndi
import pylab as pl
%pylab inline
```

```python
im1 = ndi.imread("images.jpg")
```

```python
im2 = np.dstack([im1[:,:,i].astype(float)/255 for i in range(3)])
```

## color balance

```python
pl.figure()
pl.imshow(im2)
pl.figure()
im3 = (im2 * np.atleast_2d([1.25, 0.75, 1.25])).clip(0,1)
pl.imshow(im3)
```

## gamma correction (contrast)

```python
fig = pl.figure(131)
# gamma correction to change contrast
# a power < 1 (here 0.2) lowers the contrast
imgsq = im1.astype(float)**0.5
#normalize
imgsq *= 255 / imgsq.max()
ax=fig.add_subplot(131)
pl.imshow(imgsq.astype(np.uint8))
pl.title("low contrast")

ax=fig.add_subplot(132)
pl.imshow(im1)
pl.title("original")

imgsq = im1.astype(float)**2
# a power > 1 (here 2) highers the contrast
imgsq *= 255 / imgsq.max()
ax=fig.add_subplot(133)
pl.imshow(imgsq.astype(np.uint8))
title = pl.title("high contrast")
```

## resolution 

```python
imglr = ndi.filters.uniform_filter(im1, [5,5,0])
pl.imshow(imglr)
```
